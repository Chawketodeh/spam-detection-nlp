```python
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
import pandas as pd # Import pandas if not already imported

# Download French stopwords if not already downloaded
nltk.download('stopwords')

# Ensure columns are string type to handle potential non-string data
df['dep_name_str'] = df['dep_name'].astype(str)
df['region_name_str'] = df['region_name'].astype(str)


# Conversion en minuscules
df['dep_name_lower'] = df['dep_name_str'].str.lower()
df['region_name_lower'] = df['region_name_str'].str.lower()

# Tokenization
df['dep_name_tokens'] = df['dep_name_lower'].apply(word_tokenize)
df['region_name_tokens'] = df['region_name_lower'].apply(word_tokenize)

# Suppression de la ponctuation
df['dep_name_tokens_no_punct'] = df['dep_name_tokens'].apply(lambda tokens: [word for word in tokens if word not in string.punctuation])
df['region_name_tokens_no_punct'] = df['region_name_tokens'].apply(lambda tokens: [word for word in tokens if word not in string.punctuation])


# Charger les stopwords français
french_stopwords = set(stopwords.words('french'))
# Suppression des stopwords pour les tokens de département et région
df['dep_name_tokens_no_stopwords'] = df['dep_name_tokens_no_punct'].apply(lambda tokens: [word for word in tokens if word not in french_stopwords])
df['region_name_tokens_no_stopwords'] = df['region_name_tokens_no_punct'].apply(lambda tokens: [word for word in tokens if word not in french_stopwords])

# Vérification
display(df.head())
```
